```python
def map_value(x):
    if x > 0.7:
        return 1
    elif x < 0.3:
        return 0
    else:
        return 'risk'


_list_predictions_cj1 = []
_list_y_cj1 = []
for i in range(len(X_test)):
    if(X_test[i,1] == 1):
        _list_y_cj1.append(y_test[i])
        _list_predictions_cj1.append(predictions[i,0])


print((len(_list_predictions_cj1), len(_list_y_cj1)))
print(_list_predictions_cj1)
print("CJ1 y:")  
print("CJ1 predictions:")
good_look = list(map(map_value, _list_predictions_cj1))   
print(", ".join(map(str, _list_y_cj1)))
print(", ".join(map(str, good_look)))   
```

```python
def calculate_accuracy(arr1, arr2):
    if len(arr1) != len(arr2):
        raise ValueError("Arrays must have the same length")
    
    correct_predictions = sum(int(a) == b for a, b in zip(arr1, arr2))
    accuracy = correct_predictions / len(arr1)
    return accuracy

# Calculate accuracy
accuracy = calculate_accuracy(_list_y_cj1, good_look)
print(f"Accuracy: {accuracy:.2f}")
```
